```python
%%time

model = LightFM(random_state=0,
                loss='warp',
                learning_rate=0.03,
                no_components=100)

model = model.fit(clientes_interaction_csr,
                  epochs=100,
                  num_threads=16, verbose=False)
```

Y ahora definimos la funcion para obtener recomendaciones para un usuario en espec√≠fico:

```python
def recomm(model, clientes, user_id, user_dict, item_dict, nrec_items = 20, show = True):

    n_users, n_items = clientes.shape
    user_x = user_dict[user_id]

    scores = pd.Series(model.predict(user_x,np.arange(n_items)))
    scores.index = clientes.columns
    scores = list(pd.Series(scores.sort_values(ascending=False).index))
    
    known_items = list(pd.Series(clientes.loc[user_id,:] \
                                [clientes.loc[user_id,:] > 0].index).sort_values(ascending=False))
    
    #Filtro de items que ya ley√≥ el usuario
    scores = [x for x in scores if x not in known_items]
    recomms = scores[0:nrec_items]

    print(recomms)

    if show == True:
        print ("User: " + str(user_id))
        print("Known Likes:")
        counter = 1
        for i in known_items:
            print(str(counter) + '- ' + str(i))
            counter+=1

        print("\n Recommended Items:")
        
        counter = 1
        for i in recomms:
            print(str(counter) + '- ' + str(i))
            counter+=1
    return recomms
```

Al poner la opci√≥n show=True nos imprime los known likes y recomms.

```python
recomms_test = recomm(model, clientes_interaction ,10 , user_dict, item_dict, show=True)
```

#Metricas

Para usuarios cold start (que no aparecen en el set de train), debemos buscar alguna alternativa. En este caso les recomendaremos los videos con rating promedio m√°s alto.

```python
cold_start_recomms = clientes_train\
                      .groupby("asset_id")\
                      .agg({"rating": "mean"})\
                      .sort_values(by="rating", ascending=False)\
                      .index\
                      .to_list()
```

A todos los usuarios nuevos le vamos a recomentdar los cold_start

Vamos a pasar los asset_id recomendados a los content_id que es lo que en realidad queremos recomendar

```python
cold_start_df = pd.DataFrame(cold_start_recomms, columns=['asset_id'])
cold_start_content = cold_start_df.merge(meta, how='left', left_on='asset_id', right_on='asset_id')['content_id']    
cold_start_content = cold_start_content.dropna().unique()
series_cold_start = cold_start_content.squeeze()
```

```python
series_cold_start
```

Ahora generemos recomendaciones para todos los usuarios.

```python
users = clientes_res.account_id.unique()
non_cold_users = clientes_train.account_id.unique()

print(len(users), len(non_cold_users))
```

```python
%%time

recomms_dict = {
    "user": [],
    "recomms": [],
    "cold": []
}

for user in users:
  
  # Recomms personalizadas
  if user in non_cold_users:
    user_recomms = recomm(model, clientes_interaction, user, user_dict, item_dict, nrec_items = 20, show =False)
    recomms_dict["user"].append(user)
    # En la linea siguiente pasamamos el codigo recomendado de 'asset_id' a 'content_id' 
    content_id = meta.loc[meta.asset_id.isin(user_recomms), "content_id"].unique().tolist()
    recomms_dict["recomms"].append(content_id)
    recomms_dict["cold"].append(False) #Para luego tener identificados a los usuarios cold
  
  # Recomms cold
  else:
    recomms_dict["user"].append(user)
    recomms_dict["recomms"].append(series_cold_start)
    recomms_dict["cold"].append(True) #Para luego tener identificados a los usuarios cold
```

Ahora tenemos un diccionario de recomedaciones. Lo vamos a convertir en dataframe para poder manipularlo m√°s f√°cil (como estamos acostumbrados)

Primero que nada, vamos a unir en un mismo dataframe las recomendaciones nuestras con el set ideal.

```python
df_map = ideal_recomms.merge(recomms_df, how="left", left_on="account_id", right_on="user")[["account_id", "content_id", "recomms"]]
df_map.columns = ["account_id", "ideal", "recomms"]
df_map.head()
```

```python
aps = [] # lista vac√≠a para ir almacenando la AP de cada recomendaci√≥n

for pred, label in df_map[["ideal", "recomms"]].values:
  n = len(pred) # cantidad de elementos recomendados
  arange = np.arange(n, dtype=np.int32) + 1. # indexamos en base 1 
  rel_k = np.in1d(pred[:n], label) # lista de booleanos que indican la relevancia de cada √≠tem
  tp = np.ones(rel_k.sum(), dtype=np.int32).cumsum() # lista con el contador de verdaderos positivos
  denom = arange[rel_k] # posiciones donde se encuentran los √≠tems relantes
  ap = (tp / denom).sum() / len(label) # average precision
  aps.append(ap)
```

```python
MAP = np.mean(aps)
print(f'mean average precision = {round(MAP, 5)}')
```
